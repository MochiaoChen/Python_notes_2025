```python
def is_even(num):
    '''Returns True if num is even(偶数), False otherwise.'''
    return num % 2 == 0

def main():
    number = int(input("Enter a number: "))
    if is_even(number):
        print("Even")
    else:
        print("Odd")

main()

```
